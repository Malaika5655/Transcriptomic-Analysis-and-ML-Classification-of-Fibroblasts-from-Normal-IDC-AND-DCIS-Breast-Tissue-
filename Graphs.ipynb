```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a pandas DataFrame
file_path = "/content/BP clean.xlsx"
df = pd.read_csv(file_path)

# --- 1. Bar Graph: Top 5 'Terms' by 'Fold Enrichment' ---

# Data Cleaning: Drop rows where 'Terms' is missing
df_cleaned_bar = df.dropna(subset=['Terms'])

# Sort by 'Fold Enrichment' in descending order and take the top 5
df_top_fold_enrichment = df_cleaned_bar.sort_values(by='Fold Enrichment', ascending=False).head(5)

# Create the bar plot
plt.figure(figsize=(10, 7))
sns.barplot(x='Fold Enrichment', y='Terms', data=df_top_fold_enrichment, palette='magma')
plt.xlabel('Fold Enrichment', fontsize=12)
plt.ylabel('Terms', fontsize=12)
plt.title('Top 5 Terms by Fold Enrichment', fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('bar_terms_fold_enrichment.png')
# plt.show() # Optional
print("Bar graph 'bar_terms_fold_enrichment.png' created for 'Terms' vs 'Fold Enrichment'.")

# --- 2. Histogram: Distribution of 'Pvalue' ---

# Create the histogram
plt.figure(figsize=(8, 6))
# Drop NaN Pvalues if any, though from previous info, Pvalue was non-null. Still good practice.
sns.histplot(df['Pvalue'].dropna(), kde=True, bins=10, color='skyblue')
plt.xlabel('Pvalue', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of Pvalue Distribution', fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('histogram_pvalue.png')
# plt.show() # Optional
print("Histogram 'histogram_pvalue.png' created for the 'Pvalue' column.")

# --- 3. Bar Graph: 'Category' vs. average 'Count' ---
# Since 'Category' has only one value, a bar graph showing its average count
# might not be very insightful directly, but we can still generate it.
# Alternatively, we can pick another set of columns for a more interesting bar graph.
# Let's try 'Terms' vs 'Pvalue' for the top 5 terms with lowest Pvalue (most significant)

df_cleaned_pvalue_bar = df.dropna(subset=['Terms'])
df_top_pvalue_terms = df_cleaned_pvalue_bar.sort_values(by='Pvalue', ascending=True).head(5)

plt.figure(figsize=(10,7))
sns.barplot(x='Pvalue', y='Terms', data=df_top_pvalue_terms, palette='coolwarm')
plt.xlabel('Pvalue (lower is more significant)', fontsize=12)
plt.ylabel('Terms', fontsize=12)
plt.title('Top 5 Terms by Lowest Pvalue', fontsize=15)
plt.xscale('log') # Pvalues can span orders of magnitude, log scale is often better
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('bar_terms_pvalue.png')
print("Bar graph 'bar_terms_pvalue.png' created for top 5 'Terms' vs 'Pvalue'.")

# --- 4. Histogram: Distribution of 'Fold Enrichment' ---
plt.figure(figsize=(8, 6))
sns.histplot(df['Fold Enrichment'].dropna(), kde=True, bins=10, color='salmon')
plt.xlabel('Fold Enrichment', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of Fold Enrichment Distribution', fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('histogram_fold_enrichment.png')
print("Histogram 'histogram_fold_enrichment.png' created for the 'Fold Enrichment' column.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file into a pandas DataFrame
# Use the path from your error message and pd.read_excel()
file_path = "/content/BP clean.xlsx"
try:
    # If your Excel file has multiple sheets, and the data isn't in the first one,
    # you might need to specify the sheet name, e.g., df = pd.read_excel(file_path, sheet_name="Sheet1")
    df = pd.read_excel(file_path)
    print("Excel file loaded successfully!")
except FileNotFoundError:
    print(f"Error: The file was not found at '{file_path}'")
    print("Please ensure the file path is correct and the file exists in that location.")
    # Exit if file not found, or handle appropriately
    exit()
except Exception as e:
    print(f"An error occurred while reading the Excel file: {e}")
    # Exit or handle appropriately
    exit()


# --- 1. Bar Graph: Top 5 'Terms' by 'Fold Enrichment' ---
if 'Terms' in df.columns and 'Fold Enrichment' in df.columns:
    # Data Cleaning: Drop rows where 'Terms' is missing
    df_cleaned_bar = df.dropna(subset=['Terms'])

    # Sort by 'Fold Enrichment' in descending order and take the top 5
    df_top_fold_enrichment = df_cleaned_bar.sort_values(by='Fold Enrichment', ascending=False).head(5)

    if not df_top_fold_enrichment.empty:
        plt.figure(figsize=(10, 7))
        sns.barplot(x='Fold Enrichment', y='Terms', data=df_top_fold_enrichment, palette='magma')
        plt.xlabel('Fold Enrichment', fontsize=12)
        plt.ylabel('Terms', fontsize=12)
        plt.title('Top 5 Terms by Fold Enrichment', fontsize=15)
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=10)
        plt.tight_layout()
        plt.savefig('bar_terms_fold_enrichment.png')
        print("Bar graph 'bar_terms_fold_enrichment.png' created for 'Terms' vs 'Fold Enrichment'.")
    else:
        print("Could not generate bar graph for 'Terms' vs 'Fold Enrichment': No data after filtering.")
else:
    print("Columns 'Terms' or 'Fold Enrichment' not found. Skipping bar graph 1.")


# --- 2. Histogram: Distribution of 'Pvalue' ---
if 'Pvalue' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df['Pvalue'].dropna(), kde=True, bins=10, color='skyblue')
    plt.xlabel('Pvalue', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title('Histogram of Pvalue Distribution', fontsize=15)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.savefig('histogram_pvalue.png')
    print("Histogram 'histogram_pvalue.png' created for the 'Pvalue' column.")
else:
    print("Column 'Pvalue' not found. Skipping histogram 1.")


# --- 3. Bar Graph: Top 5 'Terms' by lowest 'Pvalue' ---
if 'Terms' in df.columns and 'Pvalue' in df.columns:
    df_cleaned_pvalue_bar = df.dropna(subset=['Terms'])
    df_top_pvalue_terms = df_cleaned_pvalue_bar.sort_values(by='Pvalue', ascending=True).head(5)

    if not df_top_pvalue_terms.empty:
        plt.figure(figsize=(10,7))
        sns.barplot(x='Pvalue', y='Terms', data=df_top_pvalue_terms, palette='coolwarm')
        plt.xlabel('Pvalue (lower is more significant)', fontsize=12)
        plt.ylabel('Terms', fontsize=12)
        plt.title('Top 5 Terms by Lowest Pvalue', fontsize=15)
        # Pvalues can span orders of magnitude, log scale is often better if values are very small
        # Check if Pvalues are positive before applying log scale
        if (df_top_pvalue_terms['Pvalue'] > 0).all():
            plt.xscale('log')
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=10)
        plt.tight_layout()
        plt.savefig('bar_terms_pvalue.png')
        print("Bar graph 'bar_terms_pvalue.png' created for top 5 'Terms' vs 'Pvalue'.")
    else:
        print("Could not generate bar graph for 'Terms' vs 'Pvalue': No data after filtering.")
else:
    print("Columns 'Terms' or 'Pvalue' not found. Skipping bar graph 2.")

# --- 4. Histogram: Distribution of 'Fold Enrichment' ---
if 'Fold Enrichment' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df['Fold Enrichment'].dropna(), kde=True, bins=10, color='salmon')
    plt.xlabel('Fold Enrichment', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title('Histogram of Fold Enrichment Distribution', fontsize=15)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.savefig('histogram_fold_enrichment.png')
    print("Histogram 'histogram_fold_enrichment.png' created for the 'Fold Enrichment' column.")
else:
    print("Column 'Fold Enrichment' not found. Skipping histogram 2.")

# plt.show() # Uncomment this if you want to display plots directly in your environment
```
